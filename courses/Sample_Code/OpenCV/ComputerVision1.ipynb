# OpenCV uses BGR format, NOT RGB.
Image[: , : ,0] → Blue (B)
Image[: , : ,1] → Green (G)
Image[: , : ,2] → Red (R)

```python
#checking the min max range for 3 layers
print("Blue min/max:", ej_logo_img[:, :, 0].min(), ej_logo_img[:, :, 0].max())
print("Green min/max:", ej_logo_img[:, :, 1].min(), ej_logo_img[:, :, 1].max())
print("Red min/max:", ej_logo_img[:, :, 2].min(), ej_logo_img[:, :, 2].max())
```

```python
#now we will take the color image and remove some RGB Component just to see the behavior of image
#now from the color image i am going to remove Blue component
# Make a copy of original image
img_no_blue = ej_logo_img.copy()
# Set Blue channel to 0
img_no_blue[:, :, 0] = 0
# Display result
cv2_imshow(img_no_blue)
```

```python
#now we will take the color image and remove some RGB Component just to see the behavior of image
#now from the color image i am going to remove Blue component
# Make a copy of original image
img_no_green = ej_logo_img.copy()
# Set Green channel to 0
img_no_green[:, :, 1] = 0
# Display result
cv2_imshow(img_no_green)
```

```python
#now we will take the color image and remove some RGB Component just to see the behavior of image
#now from the color image i am going to remove Red component
# Make a copy of original image
img_no_red = ej_logo_img.copy()
# Set Red channel to 0
img_no_red[:, :, 2] = 0
# Display result
cv2_imshow(img_no_red)


```

```python
new_image=np.hstack((img_no_green,img_no_blue,img_no_red))
cv2_imshow(new_image)
```

<h1>How to crop the Image?</h1>
<h2>How it Works?</h2>

(0, 0) → top-left corner
First value = vertical position (y)
Second value = horizontal position (x)
So img[100:300, 200:400] means<br>
Start at y = 100, end at y = 300<br>
Start at x = 200, end at x = 400<br>
Visual Idea<br>
image.png

```python
ej_logo_img=cv2.imread("/content/drive/MyDrive/sample_data/logo.jpg")
ej_logo_img_cropped=ej_logo_img[100:300, 200:400]
cv2_imshow(ej_logo_img_cropped)
```

```python
#center cropping
h, w = ej_logo_img.shape[:2]

center_x, center_y = w // 2, h // 2

# Crop 200x200 from center
crop = ej_logo_img[center_y-100:center_y+100, center_x-100:center_x+100]
cv2_imshow(crop)
```

<h1>How to Draw a circle or rectangle around the image?<h1>




image.png

```python
img_black=np.zeros((500,500,3))
cv2.rectangle(img_black,pt1=(100,100),pt2=(300,300),color=(10,100,00),thickness=2) # put thikness=-1 if you want to fill it completely
cv2_imshow(img_black)
```

```python
#now we will draw a circle
img_black=np.zeros((500,500,3))
cv2.circle(img_black,center=(200,400),radius=50,color=(0,0,250),thickness=2) #200 → x-coordinate (horizontal position) 400 → y-coordinate (vertical position)
cv2_imshow(img_black)
```

```python
#draw a straight line
img_black=np.zeros((500,500,3))
# Draw a line from (50, 50) to (400, 400)
cv2.line(img_black, pt1=(50, 50), pt2=(400, 400),color=(0, 255, 0),thickness=3)
cv2_imshow(img_black)
```

```python
#draw a straight line
#syntax : cv2.putText(image, text, org, fontFace, fontScale, color, thickness)
img_black=np.zeros((500,500,3))

cv2.putText(
    img_black,
    text="eJagruti",
    org=(100, 250),
    fontFace=cv2.FONT_HERSHEY_SIMPLEX,
    fontScale=2,
    color=(0, 255, 0),
    thickness=3,
    lineType=cv2.LINE_AA
)
cv2_imshow(img_black)
```

Filesize difference between color and grayscale image

```python
import cv2
import os

image_path = "/content/drive/MyDrive/sample_data/logo.jpg"

img_color = cv2.imread(image_path, cv2.IMREAD_COLOR)
img_gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if img_color is None or img_gray is None:
    print("Error: Could not read image. Check the file path.")
else:
    print("Color shape:", img_color.shape)
    print("Gray shape:", img_gray.shape)

    print("Color image memory size:", img_color.nbytes, "bytes")
    print("Gray image memory size:", img_gray.nbytes, "bytes")

    print("Original file size on disk:", os.path.getsize(image_path), "bytes")
```

Simple Compression in various format

```python
import cv2
import os
from google.colab.patches import cv2_imshow
image_path = "/content/drive/MyDrive/sample_data/logo.jpg"
img = cv2.imread(image_path)

# Save as JPEG
cv2.imwrite("/content/drive/MyDrive/sample_data/test.jpg", img, [cv2.IMWRITE_JPEG_QUALITY, 85])

# Save as PNG
cv2.imwrite("/content/drive/MyDrive/sample_data/test.png", img, [cv2.IMWRITE_PNG_COMPRESSION, 9])

# Save as WebP
cv2.imwrite("/content/drive/MyDrive/sample_data/test.webp", img, [cv2.IMWRITE_WEBP_QUALITY, 80])

print("Original:", os.path.getsize(image_path), "bytes")
print("JPEG:", os.path.getsize("/content/drive/MyDrive/sample_data/test.jpg"), "bytes")
print("PNG:", os.path.getsize("/content/drive/MyDrive/sample_data/test.png"), "bytes")
print("WebP:", os.path.getsize("/content/drive/MyDrive/sample_data/test.webp"), "bytes")
img=cv2.imread("/content/drive/MyDrive/sample_data/test.webp")
cv2_imshow(img)
```

<h1>How to Capture the Videos?</h1>

```python
import cv2
import numpy as np

vid_cap=cv2.VideoCapture(0) #0 is the index first preference to inbuilt camera
while True:
  ret_bool,frame=vid_cap.read() # ret_bool is a boolean value which indicates status of camera
  cv2.imshow("webcam",frame)
  if cv2.waitKey(1) & 0xFF == ord('x'):
    break
```
